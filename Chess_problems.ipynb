# Problem 1: Given a chess board of dimension m * n. Find number of possible moves where knight can be moved on a chessboard from given position.

**Output**: All the possible moves

# Problem 2: On a chess board, Queen can move along any direction. Suppose a knight is to be placed on the board, find whether the position of knight is safe or not.

**Inputs**:
1. Queen's row and column
2. Knight's row and column

**Output**: It's safe or not

image.png

```python
# Enter Queen position row and column values
Qr,Qc = int(input("enter Queen row number: ")), int(input("enter Queen column number: "))

# Enter Knight position row and column values
Kr,Kc = int(input("\nenter Knight row number: ")), int(input("enter Knight column number: "))

print(f"\nposition of Queen in the board {Qr,Qc}")
print(f"position of Knight in the board {Kr,Kc}")

# Using abs function, we can convert negative values to positive values.
row_diff = abs(Qr-Kr)
column_diff = abs(Qc-Kc)

if (Qr == Kr or Qc == Kc or row_diff == column_diff):
  print("\nknight is not safe")
else:
  print("\nknight is safe")
```
